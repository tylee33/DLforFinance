```python
from pandas_datareader.data import DataReader
from datetime import date
series_code = 'T10Y2Y'
data_source = 'fred'
start = date(1962, 1, 1)
data = DataReader(series_code, data_source, start)
data.info()
```

```python
series_name = '10-year Treasury'
data = data.rename(columns={series_code: series_name})
data.plot(title = series_name)
plt.show()
```

```python
start = date(2000, 1, 1)
series = 'DCOILWTICO'
oil = DataReader(series, 'fred', start)
ticker = 'XOM'
stock = web.DataReader(ticker, 'yahoo', start)
data = pd.concat([stock[['Close']],oil], axis=1)
data.info()
```

```python
data.columns = ['Exxon','Oil Price']
data.plot()
plt.show()
```
