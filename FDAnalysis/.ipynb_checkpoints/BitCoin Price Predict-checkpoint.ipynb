```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from fbprophet import Prophet

plt.style.use('ggplot')
%pylab inline
pylab.rcParams['figure.figsize'] = (10, 6)

pd.options.display.float_format = '${:,.2f}'.format
```

```python
dfs = pd.read_html("https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20130428&end=20180702")
df = dfs[0]
df.head()
```

```python
df = df.iloc[::-1] #뒤집기
df.head()
```

```python
df['ds'] = df['Date']
df['y'] = df['Close**']
forecast_data = df[['ds','y']].copy()

forecast_data.head()
```

```python
m = Prophet()
m.fit(forecast_data);

future = m.make_future_dataframe(periods=332)
future.tail()

forecast = m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
plt.figure(figsize=(12,8))
plt.plot(df['y'])
plt.show()
```

```python
m.plot(forecast)
```

```python
m.plot_components(forecast);
```
